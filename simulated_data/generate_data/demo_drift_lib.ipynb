# Demo the drift library

Use this script to plot drift with various parameters.

```python
"""Imports."""

from matplotlib import pyplot as plt
import drift_lib
```

```python
drift = drift_lib.Drift(
    duration=1800,
    drift_rate=0.1,
    random_walk_sigma=2.,
    random_jump_rate=0.01,
    random_seed=1,
)

_, ax = plt.subplots(figsize=(6, 4))
ax.plot(drift.drift_times, drift.drift_vector)
ax.set_xlabel('Time (s)')
ax.set_ylabel('Drift ($\mu$m)')
plt.show()
```
